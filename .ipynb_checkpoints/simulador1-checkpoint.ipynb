```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

Se crean las variables para la ventilación espontánea

```python
vtidal = 500 #mL
frec_resp = 12 #rpm
compliance = 0.2 #L/cmH2O
resistance = 2.5 #cmH2O/L/s
t_muestreo = 60 #s
duracion_ciclo = 60/frec_resp
```

Se crea el simulador de ventilación espontánea (volumen por encima de CRF)

```python
t = np.arange(0,t_muestreo,0.01)
vol = np.sin((t*np.pi*2/duracion_ciclo)-(np.pi/2)) * vtidal/2 + vtidal/2
plt.plot(t,vol)
plt.grid(True)
plt.ylabel('Volumen')
plt.xlabel('Tiempo (segundos)')
plt.show()
```
