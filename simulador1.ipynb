```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

Se crean variables, fijadas arbitrariamente en un valor normal, para la ventilación espontánea

```python
vtidal = 500 #mL
frec_resp = 12 #rpm
compliance = 0.2 #L/cmH2O
resistance = 2.5 #cmH2O/L/s
t_muestreo = 30 #s
duracion_ciclo = 60/frec_resp
```

Se crea el simulador de ventilación espontánea (volumen por encima de CRF, presión transpulmonar)

```python
t = np.arange(0,t_muestreo,0.01)
vol = np.sin((t*np.pi*2/duracion_ciclo)-(np.pi/2)) * vtidal/2 + vtidal/2
pres = vol/(compliance * 1000)

flujo = np.array(t)
#for momento in t:
    
```

```python
figure, (vol_t, pres_t, flujo_t) = plt.subplots(3)
#fig.subptitle('Curvas volumen-tiempo, presión-tiempo')

vol_t.plot(t,vol)
vol_t.grid(True)
vol_t.set_title('Volumen contra tiempo')
#vol_t.ylabel('Volumen sobre CRF')
#vol_t.xlabel('Tiempo (segundos)')

pres_t.plot(t,pres)
pres_t.grid(True)
pres_t.set_title('Presión contra tiempo')
#pres_t.ylabel('Presión (cmH2O)')
#pres_t.xlabel('Tiempo (segundos)')

flujo_t.plot(t,flujo)
flujo_t.grid(True)
flujo_t.set_title('Flujo contra tiempo')
#flujo_t.ylabel('Presión (cmH2O)')
#flujo_t.xlabel('Tiempo (segundos)')

figure.show()
```
